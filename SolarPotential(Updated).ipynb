```python
#change column names
#renaming variables that are of primary focus
combine.rename(columns={'YYYY-MM-DD':'Date','HH:MM (LST)':'Time','METSTAT Glo (Wh/m^2)':'GHI'},inplace=True)
combine.head()
```

```python
#Make separate columns for day, month, and year
import datetime
#year
combine['Year']=pd.DatetimeIndex(combine['Date']).year
#month
combine['Month']=pd.DatetimeIndex(combine['Date']).month
#day
combine['Day']=pd.DatetimeIndex(combine['Date']).day
combine.head()
```

```python
#Focus on GHI data
#create subset containing variables Year, Month, Day, Time, GHI, City
sub1=combine[['Date','Year','Month','Day','Time','GHI','City']]
sub1.head(5)
```

```python
#First I will do summary statistics for all islands combined
#Hourly average
hourly_average=sub1['GHI'].mean()
hourly_average #204.75 which is higher that earth averageâ‰ˆ164
```

```python
#Daily average
daily_average=hourly_average*24 #since each row corresponds to a particular hour in a day
daily_average #4913.92
```

```python
#I will look at average GHI per day in each month
#Need to first calculate the total number of days in each month for 20 year span
nmonth=sub1.groupby('Month',as_index=False)['Day'].count() #Counts number of days in each month
nmonth['Day'] = nmonth['Day'].apply(lambda Day: Day/24) #each tuple represents an hour so we need to account for this 
nmonth.head()
```

```python
#Find average GHI per day in each month
day_sum=sub1.groupby('Month',as_index=False)['GHI'].sum() #Total GHI produced in each month over 20 years
day_average=(day_sum['GHI']/nmonth['Day'])/1000 #Convert to kWh/m2/day
day_average=pd.concat([nmonth['Month'],day_average ], axis=1, sort=False) #Convert to dataframe
day_average.rename(columns={0:'MeanGHI'},inplace=True) #Rename column
day_average.sort_values(by=['MeanGHI'],ascending=False).head() #Average day in June has highest GHI
```

```python
#Plot the information above
import matplotlib.pyplot as plt
plt.style.use('classic')
import seaborn as sns
sns.set()
plot1=sns.lineplot(x='Month',y='MeanGHI',data=day_average)
plot1.set_ylabel('GHI (kWh/m2/Day)')
plot1.set_title('Solar Energy in Hawaii')
plot1.set(ylim=(3.3,6),xticks=np.arange(1, 12, step=1))
```

```python
#Find average GHI for each month
month_sum=sub1.groupby('Month',as_index=False)['GHI'].sum() #Total GHI produced in each month over 20 years
month_average=month_sum['GHI']/(20*4) #Divide by 20*4 to get average month across all cities
month_average=month_average/1000 #Convert to kWh
month_average=pd.concat([month_sum['Month'],month_average], axis=1, sort=False)
month_average.rename(columns={'GHI':'MeanGHI'},inplace=True)
month_average.sort_values(by=['MeanGHI'],ascending=False).head() #July has the highest monthly average GHI 
```
